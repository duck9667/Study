# 제 1장 데이터 모델링의 이해

## 제 1절 데이터 모델의 이해

### 1. 모델링의 이해

가. 모델링의 정의 : 복잡한 현실 세계를 일정한 표기법에 의해 표현하는 일

나. 모델링의 특징 : **추상화** - 일정한 형식에 따라 표현 / **단순화** - 쉽게 이해하도록 도움 / **명확화** - 누구나 이해하도록 

다. 모델링의 세가지 관점 

1) 데이터 관점 : 업무가 어떤 데이터와 관련이 있는지 또는 데이터간의 관계는 무엇인지에 대해 모델링하는 방법(What, data)

2) 프로세스 관점 : 업무가 실제하고 있는 일은 무었인지 또는 무엇을 해야하는지 (How, process)

3) 데이터와 프로세스의 상관관점 : 업무가 처리하는 일의 방법에 따라 어떻게 영향을 받고 있는지 모델링하는 방법(Interaction)

### 2. 데이터 모델의 기본 개념의 이해

가. 데이터 모델링의 정의 : 정보시스템을 구축하기 위한 분석/설계 과정

나. 데이터 모델이 제공하는 기능 : 시스템 가시화,구조화,명세화,다양한 관점 제공

### 3. 데이터 모델링의 중요성 및 유의점

가. 파급효과 : 한번 해놓으면 편하다.

나. 복잡한 정보 요구사항의 간결한 표현(Concieseness)

다. 데이터 품질 : 중복 / 비유연성 / 비일관성

### 4. 데이터 모델링의 3단계 진행

가. 개념적 데이터 모델링 : 추상화 수준 높고 업무중심적이고 포괄적인 수준의 모델링 진행, 전사적 데이터 모델링, EA수립시 많이 이용

나. 논리적 데이터 모델링 : 시스템으로 구축하고자 하는 업무에 대해 key, 속성, 관계 등을 정확하게 표현, 재사용성이 높음**(가장 핵심적인 부분)**

다. 물리적 데이터 모델링 : 실제로 데이터베이스에 이식할 수 있도록 성능, 저장 등 물리적인 성격을 고려하여 설계

*실무에서는 개념적 데이터 모델링과 논리적 데이터 모델을 합쳐 논리적 데이터 모델링을 하는 경우가 대부분이다.

### 5. 프로젝트 생명주기에서 데이터 모델링

Waterfall 기반에서는 분석단계에서 논리적 모델링 / 설계 단계에서 물리적 모델링

나선형 모델에서는 업무크기에 따라 논리적,물리적 모델링이 분석 설계 단계 양쪽에서 수행, 비중은 분석단계에서 논리적 모델링에 둔다,

### 6. 데이터 모델링에서 데이터독립성의 이해

가. 데이터 독립성의 필요성 : 

지속적으로 증가하는 유지보수 비용을 절감하고 데이터 복잡도를 낮추며 중복된 데이터를 줄이기 위한 목적있다. 또한 끊임없이 요구되는 사용자 요구사항에 대해 화면과 데이터베이스 간에 서로 독립성을 유지하기 위한 목적으로 개념 출현

나. 데이터베이스 3단계 구조 :

Ansi/sparc의 3단계 구성의 데이터 독립성 모델은 외부단계와 개념적 단계, 내부적 단계로 구성되며 서로 간섭하지 않는 모델을 제시

다. 데이터독립성 요소 : 

1) 외부스키마 

View단계 여러개의 사용자관점으로 구성, 사용자 개개인이 보는 DB스키마

2) 개념스키마

모든 사용자 관점을 통합한 조직 전체의 DB를 기술하는것

3) 내부스키마

DB가 물리적으로 저장된 형식, 물리적 장치에 데이터가 실제적으로 저장되는 방법을 표현하는 스키마

라. 두 영역의 데이터독립성

**1) 논리적 독립성 : 개념스키마가 변경되어도 외부스키마에는 영향을 미치지 않도록하는것**

2) 물리적 독립성 : 내부스키마가 변경되어도 외부/개념스키마는 영향을 미치지 않도록하는것

*저장장치의 구조변경은 응요프로그램과 개념스키마에 영향없음

마. 사상(Mapping) : 상호독립적인 개념을 연결시켜주는 다리를 뜻함

1) 외부적/개념적 사상(논리적 사상) : 외부적 뷰와 개념적 뷰의 상호관련성을 정의

*사용자가 접근하는 형식에 따라 다른 타입의 필드

2) 개념적/내부적 사상(물리적 사상) : 개념적 뷰와 저장된 데이터베이스의 상호관련성을 정의

*저장된 DB 구조가 바뀐다면 개념적,내부적 사상이 바뀌어야한다. 그래야 개념적 스키마가 유지

### 7. 데이터 모델링의 중요한 세가지 개념

가. 데이터 모델링의 세가지요소 / 나. 단수와 집합 명명

1) 업무가 관여하는 어떤 것(Things) : 엔터티 / 인스턴스,어커런스

2) 어떤 것이 가지는 성격(Attributes) : 관계 / 패어링

3) 업무가 관여하는 어떤 것 간의 관계(Relationships) : 속성 / 속성값

### 8. 데이터 모델링의 이해관계자

가. 이해관계자의 데이터 모델링 중요성 인식 : 모든 시스템이 데이터 기반으로 구축되기 때문이다.

나. 데이터 모델링의 이해관계자 : 모두가 데이터 모델링에 대해 정확하게 알아야한다.

### 9. 데이터 모델의 표기법인 ERD(Entity relationship model)의 이해

*대표적인 데이터 모델 표기법 : IE 표기법과 Case*Method/Barker 표기법

### 10. 좋은 데이터 모델의 요소

가. 완전성 : 업무에 필요한 데이터는 모두 정의되어야 한다

나. 중복배제

다. 업무규칙 : 모두가 공유하므로 동일한 판단을 내릴 수 있도록 규칙이 필요

라. 데이터 재사용 : 통합모델이어야만 데이터재사용성을 향상시킬 수 있다. 효율성 증대 초래

마 의사소통 도구로서 역할

바. 통합성 : 전체 조직 관점에서 동일한 데이터가 여러 부서에 존재, 통합할 필요가 있다.

## 제 2절 엔터티

### 1. 엔터티의 개념

엔터티는 명사, 업무상 관리가 필요한 관심사, 저장되기 위한 어떤것

### 2. 엔터티와 인스턴스에 대한 내용과 표기법

IE 표기법 / Baker 표기법

### 3. 엔터티의 특징

엔터티는 반드시 속성이 있어야한다. 또한 영속적으로 존재하는 인스턴스의 집합이어야 한다. 즉, 한개가 아니라 두개 이상이어야 한다.

가. 업무에서 필요로 하는 정보

나. 식별이 가능해야 한다 - 인스턴스가 식별자에 의해 한 개씩만 존재해야함

다. 인스턴스의 집합 - 두개 이상이어야 집합 개념으로 엔터티 성립됨

라. 업무프로세스에 의해 이용 - 업무에 이용안되면 잘못된 엔터티이거나 프로세스 누락이 있는지 확인필요

마. 속성을 포함

바. 관계가 존재해야한다. 엔터티는 다른 엔터티와 최소 한개 이상의 관계가 존재해야 한다.



### 4. 엔터티의 분류

가. 유무형에 따른 분류

- 유형 엔터티 : 물리적인 형태가 있고 안정적, 업무로부터 엔터티를 구분하기 가장 용이 ex) 사원,물품
- 개념 엔터티 : 물리적인 형태가 없고 관리해야 할 개념적 정보로 구분 ex) 조직, 보험 상품
- 사건 엔터티 : 업무를 수행함에 따라 발생되는 엔터티로 비교적 발생량이 많으며 각종 통계 자료에 이용될 수 있다. ex) 주문, 청구

나. 발생시점에 따른 분류

- 기본/키엔터티 : 그 업무에 원래 존재하는 정보 다른 엔터티와 관계에 의해 생성되지 않고 독립적으로 생성 가능, 자신은 타 엔터티의 부모 역할 ex) 사원, 부서
- 중심엔터티 : 기본 엔터티로부터 발생되고 그 업무에 있어 중심적인 역할, 데이터의 양이 많이 발생되고 다른 엔터티와 관계를 통해 많은 행위 엔터티를 생성 ex) 계약, 청구, 주문
- 행위엔터티 : 두개 이상의 부모 엔터티로부터 발생, 자주 내용이 바뀌거나 데이터 양이 증가된다. ex) 주문 목록, 사원변경이력

### 5. 엔터티의 명명

- 가능한 현업업무에서 사용한 용어
- 가능하면 약어 사용 자제
- 단수명사 사용
- 모든 엔터티에서 유일하게 이름이 부여되어야 한다.
- 엔터티 생성의미대로 이름을 부여

## 제 3절 속성

### 1. 속성의 개념

- 업무에서 필요
- 의미상 더 이상 분리되지 않음
- 엔터티를 설명하고 인스턴스의 구성요소가 된다.

### 2. 엔터티, 인스턴스와 속성, 속성값에 대한 내용과 표기법

가. 엔터티, 인스턴스, 속성, 속성값의 관계

- 한개의 엔터티는 두 개 이상의 인스턴스의 집합
- 한개의 엔터티는 두 개 이상의 속성
- 한개의 속성은 한 개의 속성값

ex) 사원 - 엔터티 / 이름,나이 - 속성 / 홍길동, 27살 - 속성값

### 3. 속성의 특징

- 업무에 필요하고 관리하고자 하는 정보
- 정규화 이론에 근간하여 정해진 주식별자에 함수적 종속성을 가져야한다
- 하나의 속성에는 한개의 속성값, 여러 값이 있는 다중값을 경우 별도의 엔터티를 이용하여 분리

### 4. 속성의 분류

가. 속성의 특성에 따른 분류

- 기본 속성 : 업무로부터 추출한 모든 속성이 해당, 가장 일반적. 다른 속성을 계산하거나 영향을 받아 생성된 속서을 제외한 모든 속성

*주의할 점은 업무로부터 분석한 속성이라도 이미 업무상 코드로 정의한 속성이 많다. 이러한 경우도 속성값이 원래 속성을 나타내지 못하므로 기본 석성이 되지 않는다

- 설계 속성 : 업무상 필요한 데이터 이외에 데이터 모델링을 위해, 새로 만들거나 변형하여 정의한 속성 ex) 일련번호와 같은 속성은 단일한 식별자를 부여하기 위해 모델 상에서 새로 정의하는 설계 속성
- 파생 속성 : 다른 속성에 영향을 받아 발생하는 속성, 보통 계산된 값들이 이에 해당, 가급적 적게 정의하는 것이 좋다

### 5. 도메인

각 속성이 가질 수 있는 값의 범위

### 6. 속성의 명명

- 아무리 일반적이어도 업무에서 사용하지 않는다면 제외
- 서술식과 소유격은 사용하지 않는다.

## 제 4절 관계

### 1. 관계의 개념

가. 관계의 패어링 : 엔터티 안에 인스턴스가 개별적으로 관계를 가지는 것

### 2. 관계의 분류

가. 존재에 의한 관계 : 사원은 부서에 항상 속해 있다. - 연관관계

나. 행위에 의한 관계 : 주문은 고객이 주문을 할 때 발생한다. - 의존관계

*UML에는 클래스다이어그램의 관계 중 연관,의존관계가 있다. 클래스다이어그램에서는 이 둘을 구분하여 표시

### 3. 관계의 표기법

가. 관계명 : 소속,포함된다. 

*수동 능동표현으로 구분하고 현재형으로 표현한다.

나. 관계차수 : 1:1, 1:M 

다. 관계선택 사양 : 필수관계, 선택관계

## 제 5절 식별자

### 1. 식별자 개념

하나의 엔터티에 구성되어 있는 여러 개의 속성 중에 엔터티를 대표할 수 있는 속성을 의미, 하나의 엔터티는 반드시 하나의 유일한 식별자가 필요하다.

### 2. 식별자의 특징

- 유일성 : 주식별자에 의해 엔터티 내에 모든 인스턴스들이 유일하게 구분되어야한다.
- 최소성 : 주식별자를 구성하는 속성의 수는 유일성을 만족하는 최소의 수
- 불변성 : 지정된 주식별자의 값은 자주 변하지 않아야한다.
- 존재성 : 주식별자가 지정되면 반드시 값이 들어와야한다.

### 3. 식별자 분류 및 표기법

가. 식별자 분류

- 대표성 여부 : 주식별자 / 보조식별자
- 스스로 생성 여부 : 내부식별자 / 외부식별자
- 속성의 수 : 단일식별자 / 복합식별자
- 대체 여부 : 본질적식별자 / 인조식별자

### 4. 주식별자 도출기준

### 5. 식별자관계와 비식별자관계에 따른 식별자
